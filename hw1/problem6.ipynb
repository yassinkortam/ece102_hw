```python
import matplotlib.pyplot as plt
import numpy as np
import math

RANGE = 10
REFINEMENT = 100
```

Defining message and output functions

```python
#message function 1
def m(t):
    return np.cos(t)

#message function 2
def m1(t):
    if m(t) < 0:
        return -1
    else:
        return 1

#output function 1
def y1(v,t):
    return np.cos(2*np.pi*t + 2*np.pi*v*np.sin(t))

#output function 2
def y2(v,t):
    return np.cos(2*np.pi*t*(1+ v*m1(t)))
```

Defining discrete array generator

```python
#Discrete array generator converts functions to arrays
def to_arrayv(f,v):
    y = []
    for t in range(RANGE*REFINEMENT):
        y.append(f(v,t/REFINEMENT))
    return y
def to_array(f):
    y = []
    for t in range(RANGE*REFINEMENT):
        y.append(f(t/REFINEMENT))
    return y
```

Plotting function 1 when v=1 and v=10

```python
mt = to_array(m)
y1t = to_arrayv(y1,1)
y10t = to_arrayv(y1,10)
ytime = [i/REFINEMENT for i in range(RANGE*REFINEMENT)]

plt.figure()
plt.title("FM Signal m(t)")
plt.plot(ytime, mt, label="m(t)")
plt.plot(ytime, y1t, label="V=1")
plt.plot(ytime, y10t, label="V=10")
plt.legend()
```

It is clear from the graph that the FM transmitter is not a linear system, since the output is not a linear transformation of the input signal. I.e. the system does not satisfy superposition or scaling. 

Plotting function 2 when v=1 and v=10

```python
y21t = to_arrayv(y2,1)
y210t = to_arrayv(y2,10)

plt.figure()
plt.title("FM Signal m1(t)")
plt.plot(ytime, mt, label="m(t)")
plt.plot(ytime, y21t, label="V=1")
plt.plot(ytime, y210t, label="V=10")
plt.legend()
```
