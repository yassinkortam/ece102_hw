```python
import matplotlib.pyplot as plt
import numpy as np
import math

RANGE = 10
REFINEMENT = 10
```

defining ramp, unit step, and unit impulse

```python
#ramp function
def r(t):
    if t < 0:
        return 0
    else:
        return t

#unit step function
def u(t):
    if t < 0:
        return 0
    else:
        return 1

#unit impulse function
def d(t):
    if t != 0:
        return 0
    else:
        return 1
```

defining the input signals and impulse responses given 

```python
#impulse response
def h(t):
    return math.exp(-2*t)*u(t)

#input function a
def ya(t):
    return np.cos(2*np.pi*t)*(u(t)-u(t-20))

#input function b
def yb(t):
    return np.sin(np.pi*t)*math.exp(-0.1*t)*(u(t)-u(t-20))

#input function c
def yc(t):
    return r(t) - 2*r(t-2) + r(t-4)
```

Defining discrete array generator

```python
#Discrete array generator converts functions to arrays
def to_array(f):
    y = []
    for t in range(RANGE*REFINEMENT):
        y.append(f(t/REFINEMENT))
    return y

#convolution integral
def c(x,h):
    return np.convolve(x,h)
```

Plotting the impulse response and input signals

```python
ht = to_array(h)
yat = to_array(ya)
ybt = to_array(yb)
yct = to_array(yc)
ytime = [i/REFINEMENT for i in range(RANGE*REFINEMENT)]

oa = c(yat,ht)
ob = c(ybt,ht)
oc = c(yct,ht)
otime = [i/REFINEMENT for i in range(len(oa))]

plt.figure()
plt.title("Impulse Response")
plt.plot(ytime,ht)

plt.figure()
plt.title("Input A")
plt.plot(ytime,yat)

plt.figure()
plt.title("Input B")
plt.plot(ytime,ybt)

plt.figure()
plt.title("Input C")
plt.plot(ytime,yct)

plt.figure()
plt.title("Output A")
plt.plot(otime,oa)

plt.figure()
plt.title("Output B")
plt.plot(otime,ob)

plt.figure()
plt.title("Output C")
plt.plot(otime,oc)
```
